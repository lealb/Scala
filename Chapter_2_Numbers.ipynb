### Numbers

#### 2.1. Parsing a Number from a String
1. 通过to方法转换兼容的对象
2. parseInt可转换进制

```scala211
"100".toInt
"100".toDouble
"100".toFloat
```

```scala211
"1".toLong
"1".toShort
"1".toByte
```

```scala211
val b =BigInt("1")
val c:BigInt=1
b==c
```

```scala211
Integer.parseInt("10")
Integer.parseInt("10",2)
Integer.parseInt("10",8)
Integer.parseInt("16",16)
```

```scala211
implicit class StringToInt(s:String){
    def toInt(radix:Int)=Integer.parseInt(s,radix)
}
```

```scala211
"11".toInt(2)
"11".toInt(8)
"11".toInt(10)
"11".toInt(16)
```

```scala211
"we".toInt(2)
```

#### 2.5. Comparing Floating-Point Numbers
1. 创建一个方法，在精度范围内约等于

```scala211
var a=0.3
var b=0.1+0.2
a==b
```

```scala211
def ~=(x:Double,y:Double,precision:Double)={
    if((x-y).abs<precision)true else false
}
```

```scala211
~=(a,b,0.0001)
```
